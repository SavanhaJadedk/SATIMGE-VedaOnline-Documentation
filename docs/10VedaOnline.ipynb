
### 10.1 Retrieving SATIMGE Input Data

Efficient retrieval of SATIMGE input data from Veda Online requires an understanding of the platform's navigation and filtering capabilities. The steps below provide a guide for accessing and refining the required data while highlighting the key features of Veda Online. For more detailed guidance on Veda Online features, refer to the [Veda Online Documentation](https://vedaonline-documentation.readthedocs.io/en/latest/index.html).

---

#### 10.1.1 Accessing Veda Online
- Visit [Veda Online](https://vedaonline.cloud/Home.aspx), locate **public models** in the left panel, and select **brunomâ€™s models** > **SATIMGE**.
- Direct link: [SATIMGE on Veda Online](https://vedaonline.cloud/ModelPage.aspx).

---

#### 10.1.2 Overview of Veda Online Tabs

Veda Online provides several key tabs for navigating and interacting with model data:

- **Navigator**: Displays all Excel files included in the SATIMGE model. Use this tab to verify model structure and file inclusion.
- **Browse**: Offers a tabular view of all input data across the Excel files. This is a primary tab for filtering and reviewing data inputs.
- **Items Detail**: Shows the topology and input parameters for individual items, enabling detailed inspection.
- **Attribute Master**: Provides a comprehensive view of attributes for model items, including topology and parameters.
- **Run Manager**: Allows users to define and execute scenarios or cases for the SATIMGE model.
- **Results**: Enables users to analyze outputs from completed model runs.
- **Reports**: Facilitates further analysis of model outputs, including Sankey diagrams and other advanced visualizations.

---

#### 10.1.3 Using the Browse Tab

The **Browse** tab is central to data retrieval in Veda Online. It displays input data from all associated Excel files in a tabular format. Once in the SATIMGE model:
- Open the **Browse** tab from the top navigation bar.
- Apply filters to refine your view by:
  - **Scenario**
  - **Process**
  - **Commodity**
  - **Attribute**
  - **User Constraint**

---

#### 10.1.4 Filtering Options in Veda Online

1. **Filtering by Process**:
   - Use the **Process** filter to sort data by main sectors, including:
     - Residential, Commercial, Transport, Industry.
   - Additional sectors, such as Imports, Exports, Agriculture, and Power, can also be filtered.

2. **Filtering by Commodity**:
   - Target sector-specific commodities using the **Commodity** filter:
     - **Residential**: Use wildcard `R*`.
     - **Commercial**: Use `C*`.
     - **Transport**: Use `T*`.
   - For exact matches, refer to the detailed commodity codes in this documentation.

3. **Filtering by Attributes and User Constraints**:
   - **Attributes**: Filter by parameters such as **ACT_COST**, **FLO_MARK**, and **EFF**, tailored to each sector.
   - **User Constraints**: Define specific limits or requirements for the model inputs.

---

#### 10.1.5 Exporting Data

After applying filters, users can export the scenario-specific data:
- Available formats include **Excel** and **CSV**.
- Alternatively, access the **GitHub repository** for direct retrieval of input files: [SATIMGE Repository](https://github.com/brunomerven/SATIMGE).

---

#### 10.1.6 Generating Sankey Diagrams in Veda Online

The **Reports** tab supports the generation of Sankey diagrams for visualizing energy flows:
1. Open the **Reports** tab.
2. Under the **ATTRIBUTE** icon, select the desired Sankey option based on level and sector, then click **Get Data**.
3. Use the **Layout** icon to define source-target relationships and update the table.
4. Toggle between **Table**, **Sankey**, or **SankeyWithLoops** views to generate the diagram.

---

#### 10.1.7 Retrieving RES Diagrams

To retrieve RES diagrams for the SATIM model:
1. Access the SATIMGE repository: [https://github.com/brunomerven/SATIMGE](https://github.com/brunomerven/SATIMGE).
2. Locate the appropriate workbook for the sector (e.g., `vt_REGION_IND-PP`).
3. Open the sheet labeled **RES** within the workbook to view the structural layout.

---


## Accessing Detailed Descriptions in VedaOnline

For comprehensive descriptions of each technology, users may consult the *Attribute Master* tab in VedaOnline, which provides expanded definitions and specifications. This resource offers valuable insights into the technical characteristics of each technology code used within the SATIM model.

## 10.2 Veda Online SATIMGE Attributes Dictionary

| **Attribute** | **Description** | **Purpose in Model** |
|---------------|-----------------|----------------------|
| **ACT_BND**   | Activity bound for a process, indicating limits on activity. | Used to set upper/lower bounds on process activity. |
| **ACT_COST**  | Cost associated with the activity of a process. | Represents operating costs of a process. |
| **ACT_CUM**   | Cumulative activity over a period. | Tracks cumulative energy demand or supply over time. |
| **ACT_EFF**   | Efficiency of a process in converting energy. | Used to model process efficiency, affecting energy input-output. |
| **ACT_UPS**   | Activity of a process in a specific time slice. | Tracks activity in a defined time period (e.g., seasonal or daily). |
| **CAP_BND**   | Capacity bound for a process. | Sets the maximum or minimum installed capacity for technologies. |
| **COM_AGG**   | Commodity aggregation for energy flows. | Aggregates energy flows across multiple processes or technologies. |
| **COM_BNDNET**| Net commodity bounds for supply or demand. | Limits the net flow of energy or materials in/out of processes. |
| **COM_CSTNET**| Net cost of a commodity. | Tracks the net cost of commodity flows between processes. |
| **COM_CUMNET**| Cumulative net flows for a commodity. | Tracks total cumulative commodity flows across the system. |
| **COM_FR**    | Commodity flow ratio between processes. | Specifies ratios for how commodities flow between processes. |
| **COM_PEAK**  | Peak commodity demand or supply. | Sets a maximum limit on peak commodity flow. |
| **COM_PKRSV** | Peak reserve for a commodity. | Specifies reserve margins for peak commodity supply. |
| **COM_PROJ**  | Commodity projections over time. | Projects future commodity supply or demand trends. |
| **FLO_EMIS**  | Emissions related to process energy flows. | Tracks pollutant emissions based on energy flows. |
| **FLO_MARK**  | Flow markers for energy or material transfers. | Sets upper and lower bounds on energy/material flows between processes. |
| **FLO_SHAR**  | Share of flow that a process must supply. | Specifies the percentage share of energy or material flow. |
| **G_DRATE**   | Discount rate applied to future costs/benefits. | Used to discount future costs in economic calculations. |
| **G_DYEAR**   | Base year for discounting. | Specifies the base year for discount rate calculations. |
| **G_YRFR**    | Year fraction for discounting. | Fraction of a year used in discount rate calculations. |
| **IRE_FLO**   | Flow of imported/exported energy. | Tracks international energy imports/exports. |
| **IRE_PRICE** | Price of imported/exported energy. | Specifies the price for energy imports/exports. |
| **NCAP_AF**   | Availability factor for installed capacity. | Indicates the availability of installed capacity for use. |
| **NCAP_AFA**  | Annual availability factor. | Specifies the yearly availability of a process. |
| **NCAP_AFC**  | Capacity factor by season or time slice. | Adjusts availability by specific time periods. |
| **NCAP_AFX**  | External availability factor. | Availability based on external conditions (e.g., weather). |
| **NCAP_BND**  | Installed capacity bounds. | Sets upper/lower limits on installed capacity. |
| **NCAP_CHPR** | Charge rate for storage technologies. | Specifies the charging rate for energy storage technologies. |
| **NCAP_COST** | Cost of installed capacity. | Represents capital cost for installing new capacity. |
| **NCAP_DISC** | Discharge rate for storage technologies. | Specifies the discharge rate for storage technologies. |
| **NCAP_ELIFE**| Economic lifetime of installed capacity. | Determines how long a technology can be used before replacement. |
| **NCAP_FOM**  | Fixed operation and maintenance cost. | Fixed cost for operating and maintaining installed capacity. |
| **NCAP_ICOM** | Investment cost for new capacity. | Tracks the initial investment cost for new technologies. |
| **NCAP_ILED** | Lead time for new capacity installations. | Specifies time needed to install new capacity. |
| **NCAP_PASTI**| Past investment in installed capacity. | Tracks previous investments in infrastructure. |
| **NCAP_PKCNT**| Peak contribution of installed capacity. | Specifies the contribution of capacity during peak demand. |
| **NCAP_START**| Start-up cost for installed capacity. | Tracks costs associated with starting up new capacity. |
| **NCAP_TLIFE**| Technical lifetime of installed capacity. | Specifies the duration before technical degradation or replacement. |
| **PRC_ACTFLO**| Activity flows for processes. | Tracks energy/material flow for specific processes. |
| **PRC_CAPACT**| Capacity-activity ratio for processes. | Specifies the ratio between process capacity and activity level. |
| **PRC_REFIT** | Refitting costs for existing technologies. | Tracks costs for upgrading or refitting existing technologies. |
| **PRC_RESID** | Residual capacity of processes. | Tracks remaining usable capacity for processes over time. |
| **RCAP_BND**  | Residual capacity bounds. | Sets upper/lower bounds on residual capacity. |
| **RPT_OPT**   | Option reporting for optimization. | Specifies which options to report during optimization. |
| **SHAPE**     | Shape or time profile for energy demand/supply. | Defines the temporal profile for energy use across time slices. |
| **STG_EFF**   | Storage efficiency for energy storage technologies. | Specifies efficiency of energy storage processes. |
| **UC_ACT**    | Unit cost for process activities. | Tracks the cost per unit of activity for processes. |
| **UC_CAP**    | Unit cost of installed capacity. | Tracks the cost per unit of installed capacity. |
| **UC_COMNET** | Unit cost of commodity flows. | Tracks the cost per unit of commodity flows in the system. |
| **UC_COMPRD** | Unit cost of commodity production. | Tracks the cost per unit for producing energy/materials. |
| **UC_IRE**    | Unit cost for energy imports/exports. | Specifies the unit cost for cross-border energy flows. |
| **UC_NCAP**   | Unit cost of new capacity installations. | Tracks the cost for each unit of new capacity installed. |
| **UC_RHSRT**  | Unit cost for short-term reserve. | Tracks the cost for short-term energy reserves. |
| **UC_RHSRTS** | Unit cost for seasonal reserves. | Tracks the cost for maintaining seasonal reserves. |
| **VDA_CEH**   | Carbon emissions for specific processes. | Tracks carbon emissions from energy processes. |
| **VDA_EMCB**  | Emission control costs and benefits. | Tracks costs and benefits associated with emission controls. |


## 10.2. Veda Online SATIMGE User Constraints Dictionary

| **User Constraint**             | **Description**                                                                                       | **Purpose in Model**                                                                                           |
|----------------------------------|-------------------------------------------------------------------------------------------------------|-----------------------------------------------------------------------------------------------------------------|
| **U_CCS_CAPT**                   | Carbon capture and storage (CCS) capacity constraint.                                                  | Limits the maximum capacity of CCS technologies in the system.                                                   |
| **U_CCS_PWR**                    | Power output constraint for CCS processes.                                                             | Limits the power output from processes using CCS technology.                                                     |
| **U_CCS_PWRC**                   | Constraint on power generation using CCS.                                                              | Restricts power generation using CCS to a certain capacity or output.                                            |
| **U_CCS_PWRG**                   | Growth constraint for power generation using CCS.                                                      | Sets growth limits on power generation using CCS.                                                                |
| **U_CCS_STOR**                   | Storage constraint for CCS.                                                                            | Limits the storage capacity for captured carbon.                                                                 |
| **U_NH3CTL**                     | Ammonia control technology constraint.                                                                 | Places restrictions on ammonia control technologies in energy generation or industry.                            |
| **UC_Growth_Batteries**          | Growth constraint for battery technologies.                                                            | Sets a limit on the annual growth of battery storage capacity.                                                   |
| **UC_Growth_CO2EQS**             | Growth constraint on CO2 equivalents.                                                                  | Limits the growth of CO2-equivalent emissions from energy processes.                                             |
| **UC_Growth_CO2S**               | Growth constraint on CO2 sequestration.                                                                | Sets a limit on the growth of carbon sequestration technologies.                                                 |
| **UC_Growth_CSP**                | Growth constraint for concentrated solar power (CSP).                                                  | Limits the annual growth of installed capacity for CSP technologies.                                             |
| **UC_Growth_DAC**                | Growth constraint for direct air capture (DAC).                                                        | Restricts the expansion of DAC technologies used for carbon capture.                                             |
| **UC_Growth_ICPSCLE-E**          | Growth constraint for industrial coal processes with CCS.                                              | Sets growth limits on the use of industrial coal processes with CCS in the electricity sector.                   |
| **UC_Growth_IFBSCOA-E**          | Growth constraint for industrial biomass coal processes.                                               | Limits growth in industrial biomass coal processes for electricity generation.                                   |
| **UC_Growth_IMISCOA-E**          | Growth constraint for industrial miscellaneous coal processes.                                         | Restricts growth in the use of miscellaneous coal processes in the electricity sector.                           |
| **UC_Growth_INFSCOA-E**          | Growth constraint for industrial steel coal processes.                                                 | Limits the expansion of industrial steel production using coal in electricity generation.                        |
| **UC_Growth_IOTSCOA-E**          | Growth constraint for industrial oil processes.                                                        | Restricts growth in industrial oil-based processes used in electricity generation.                               |
| **UC_Growth_IOTSELC-E**          | Growth constraint for industrial electricity processes.                                                | Limits growth in electricity production by industrial processes.                                                 |
| **UC_Growth_IOTSELC-N**          | Growth constraint for industrial electricity processes (non-renewable).                                | Limits growth in non-renewable industrial electricity generation processes.                                      |
| **UC_Growth_IOTSGAS-E**          | Growth constraint for industrial gas processes.                                                        | Sets limits on the growth of industrial gas processes in electricity generation.                                 |
| **UC_Growth_IOTSOHF-E**          | Growth constraint for industrial oil heating processes.                                                | Restricts growth in the use of industrial oil for heating in the electricity sector.                             |
| **UC_Growth_Nuclear**            | Growth constraint for nuclear technologies.                                                            | Limits the expansion of nuclear power generation capacity.                                                       |
| **UC_Growth_PV**                 | Growth constraint for photovoltaic (PV) solar technologies.                                            | Limits the annual growth of PV solar capacity in the system.                                                     |
| **UC_Growth_TFHCV1ELC-N**        | Growth constraint for first-class electric vehicles (non-renewable).                                  | Restricts growth in non-renewable electric vehicles of class 1.                                                  |
| **UC_Growth_TFHCV2ELC-N**        | Growth constraint for second-class electric vehicles (non-renewable).                                 | Limits the growth of non-renewable second-class electric vehicles.                                               |
| **UC_Growth_TFHCV2HGN-N**        | Growth constraint for second-class hydrogen vehicles (non-renewable).                                 | Limits growth in the use of non-renewable hydrogen vehicles of class 2.                                          |
| **UC_Growth_TFHCV3ELC-N**        | Growth constraint for third-class electric vehicles (non-renewable).                                  | Restricts growth in the use of third-class non-renewable electric vehicles.                                      |
| **UC_Growth_TFHCV3HGN-N**        | Growth constraint for third-class hydrogen vehicles (non-renewable).                                  | Sets limits on the growth of third-class non-renewable hydrogen vehicles.                                        |
| **UC_Growth_TFHCV4ELC-N**        | Growth constraint for fourth-class electric vehicles (non-renewable).                                 | Limits the growth of fourth-class non-renewable electric vehicles.                                               |
| **UC_Growth_TFHCV4HGN-N**        | Growth constraint for fourth-class hydrogen vehicles (non-renewable).                                 | Restricts growth in the use of non-renewable hydrogen vehicles of class 4.                                       |
| **UC_Growth_TFHCV5ELC-N**        | Growth constraint for fifth-class electric vehicles (non-renewable).                                  | Limits the expansion of non-renewable fifth-class electric vehicles.                                             |
| **UC_Growth_TFHCV5HGN-N**        | Growth constraint for fifth-class hydrogen vehicles (non-renewable).                                  | Restricts growth in the use of non-renewable hydrogen vehicles of class 5.                                       |
| **UC_Growth_TFHCV6ELC-N**        | Growth constraint for sixth-class electric vehicles (non-renewable).                                  | Limits the annual growth of non-renewable sixth-class electric vehicles.                                         |
| **UC_Growth_TFHCV6HGN-N**        | Growth constraint for sixth-class hydrogen vehicles (non-renewable).                                  | Restricts growth in the use of sixth-class non-renewable hydrogen vehicles.                                      |
| **UC_Growth_TFHCV7ELC-N**        | Growth constraint for seventh-class electric vehicles (non-renewable).                                | Sets growth limits on seventh-class non-renewable electric vehicles.                                             |
| **UC_Growth_TFHCV7HGN-N**        | Growth constraint for seventh-class hydrogen vehicles (non-renewable).                                | Restricts growth in the use of non-renewable hydrogen vehicles of class 7.                                       |
| **UC_Growth_TFLCVELC-N**         | Growth constraint for light commercial vehicles (electric, non-renewable).                            | Limits the growth of non-renewable light commercial electric vehicles.                                           |
| **UC_Growth_TFLCVOGSH-N**        | Growth constraint for light commercial oil and gas (non-renewable).                                   | Restricts the growth of non-renewable oil and gas-powered light commercial vehicles.                             |
| **UC_Growth_TPPRCARELC-N**       | Growth constraint for private passenger cars (electric, non-renewable).                              | Limits the annual growth of private passenger electric vehicles (non-renewable).                                 |
| **UC_Growth_TPPRCARHGNF-N**      | Growth constraint for private passenger cars (hydrogen, non-renewable).                              | Restricts the growth of non-renewable hydrogen passenger cars.                                                   |
| **UC_Growth_TPPRCAROGSH-N**      | Growth constraint for private passenger cars (oil and gas, non-renewable).                           | Sets a limit on the growth of non-renewable oil and gas passenger cars.                                          |
| **UC_Growth_TPPRSUVELC-N**       | Growth constraint for sport utility vehicles (electric, non-renewable).                              | Limits the expansion of non-renewable electric SUVs.                                                             |
| **UC_Growth_TPPRSUVHGNF-N**      | Growth constraint for sport utility vehicles (hydrogen, non-renewable).                              | Restricts growth in the use of non-renewable hydrogen SUVs.                                                      |
| **UC_Growth_TPPUMBTELC-N**       | Growth constraint for public transportation buses (electric, non-renewable).                         | Limits the growth of non-renewable electric public buses.                                                        |
| **UC_Growth_Wind**               | Growth constraint for wind technologies.                                                             | Sets limits on the expansion of wind energy capacity.                                                            |
| **UC_Growth_XICPELC**            | Growth constraint for imported electricity.                                                          | Limits the growth of imported electricity sources.                                                               |
| **UC_Growth_XIFBELC**            | Growth constraint for imported fossil-based electricity.                                              | Sets a limit on the growth of fossil-based imported electricity.                                                 |
| **UC_Growth_XIMIELC**            | Growth constraint for imported mixed electricity.                                                     | Limits the growth of electricity imports from mixed sources (renewable and non-renewable).                       |
| **UC_Growth_XINDELC**            | Growth constraint for industrial electricity.                                                         | Restricts the growth of industrial electricity sources.                                                          |
| **UC_Growth_XINDHGN**            | Growth constraint for industrial hydrogen.                                                           | Sets a limit on the growth of hydrogen used in industrial applications.                                          |
| **UC_Growth_XTRAELC**            | Growth constraint for electricity transmission.                                                       | Limits the growth of transmission capacity for electricity.                                                      |
| **UCACT_CCSInd**                 | Activity constraint for CCS in the industrial sector.                                                 | Limits the activity of carbon capture and storage in industrial processes.                                       |
| **UCACT_CCSPower**               | Activity constraint for CCS in the power sector.                                                      | Sets a limit on the activity of CCS in power generation.                                                         |
| **UCACT_IPP_BIO**                | Activity constraint for independent power producers using biomass.                                   | Restricts the activity of biomass-based independent power producers.                                             |
| **UCACT_IPP_PULP**               | Activity constraint for independent power producers using pulp and paper waste.                      | Limits the activity of pulp and paper-based power producers.                                                     |
| **UCACT_IPP_RECMAX**             | Activity constraint for renewable energy capacity.                                                    | Sets a limit on the maximum activity of renewable energy independent power producers.                            |
| **UCACT_SteelScrap**             | Activity constraint for steel production using scrap metal.                                           | Restricts activity related to steel production using scrap.                                                      |
| **UCACT_WIND**                   | Activity constraint for wind energy generation.                                                       | Limits the activity of wind energy generation processes.                                                         |
| **UCCAP_BATTR**                  | Capacity constraint for battery technologies.                                                         | Sets a maximum or minimum bound on installed capacity for battery storage technologies.                          |
| **UCCAP_DUMETR**                 | Capacity constraint for metering processes.                                                           | Limits capacity for energy metering infrastructure.                                                              |
| **UCCAP_WIND**                   | Capacity constraint for wind energy technologies.                                                     | Sets a maximum or minimum bound on installed capacity for wind energy.                                           |
| **UCNCAP_REAF**                  | Capacity constraint for renewable energy access funds.                                                | Restricts the installed capacity for technologies funded through renewable energy access programs.               |


##Editors Notes:
